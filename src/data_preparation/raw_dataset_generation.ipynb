```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Определение сетки и параметров
x, y = np.meshgrid(np.arange(-2.5, 2.5, 0.01), np.arange(-2.5, 2.5, 0.01))
k = 20
z = np.arctan2(y, x)
```

```python
save_dir = '../../data/raw'
M = np.arange(0, 10, 2) #заряды вортекса 

for m in M:
    if (m % 4) != 0:
        H = - 2 * (1 + np.cos(k * x - m * z)) #инверсия цвета для зарядов не кратных 4
    else: 
         H = 2 * (1 + np.cos(k * x - m * z))
    
    plt.figure(12, figsize=(1.6, 1.6))
    plt.imshow(H, cmap='Greys', origin='lower', extent=[-2.5, 2.5, -2.5, 2.5])
    ax = plt.gca()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    #plt.savefig(os.path.join(save_dir, f'{m}.jpeg'), bbox_inches='tight')  
    plt.show()
```
